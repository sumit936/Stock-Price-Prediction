#### 4.1

```python
data = pd.read_csv("week3.csv")
data.head()
```

```python
data['Call'] = pd.DataFrame([0]*data.shape[0])
data
```

```python
print(data.lower_band.isnull().sum())
```

```python
for i in range(14,data.shape[0]):
    if data['Close Price'][i]< data.lower_band[i]:
        data.Call[i] = 'Buy'
    elif (data['Close Price'][i]> data.lower_band[i]) and (data['Close Price'][i]<data.Average[i]):
        data.Call[i] = 'Hold Buy/ Liquidate Short'
    elif (data['Close Price'][i]> data.Average[i]) and (data['Close Price'][i]<data.upper_band[i]):
        data.Call[i] = 'Hold Short/ Liquidate Buy'
    elif data['Close Price'][i]> data.upper_band[i]:
        data.Call[i] = 'Short'
```

```python
data.Average.isnull().sum()
```

```python
data.Average[14]
```

```python
data.tail()
```

### Fitting the Classifier

```python
from sklearn import tree
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score
```

```python
x = pd.concat([data.iloc[14:,20:21],data.iloc[14:,22:24],data["Close Price"][14:]],axis = 1).reset_index(drop = True)
x.isnull().sum()
```

```python
x
```

```python
y = data.Call[14:].reset_index(drop = True)
```

```python
y
```

```python
#splittin the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .20)
```

```python
dec_model = tree.DecisionTreeClassifier()
dec_model.fit(x_train,y_train)
```

```python
y_pred = dec_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
svc_model = SVC(kernel = 'poly')
svc_model.fit(x_train,y_train)
```

```python
y_pred_svc = svc_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_svc)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
y_pred_log = log_model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_log)
```

```python
pvr["perc_open_close"]  = pd.DataFrame([0]*pvr.shape[0])
pvr["perc_low_high"]  = pd.DataFrame([0]*pvr.shape[0])
for i in range(pvr.shape[0]):
    pvr["perc_open_close"][i] =  ((pvr['Open Price'][i]-pvr["Close Price"][i])/pvr['Open Price'][i])*100
    pvr["perc_low_high"][i] = ((pvr['Low Price'][i]-pvr["High Price"][i])/pvr['Low Price'][i])*100
```

```python
pvr['perc_open_close'].unique()
```

```python
pvr["Day_Perc_Change"] = pvr['Close Price'].pct_change()*100
pvr["Day_Perc_Change"] = pvr["Day_Perc_Change"].fillna(0)
pvr.head(100)
```

```python
pvr["R_mean"] = pvr["Day_Perc_Change"].rolling(5).mean()
pvr["R_std"] = pvr["Day_Perc_Change"].rolling(5).std()
```

```python
pvr
```

```python
pvr['Action'] = pd.DataFrame([0]*pvr.shape[0])
for i in range(pvr.shape[0]-1):
    if pvr["Close Price"][i+1] > pvr["Close Price"][i]:
        pvr["Action"][i] = 1
    if pvr["Close Price"][i+1] < pvr["Close Price"][i]:
        pvr["Action"][i] = -1
```
